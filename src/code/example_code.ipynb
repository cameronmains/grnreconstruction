```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# Example data
your_results = np.array([0.7, 0.8, 0.75, 0.65, 0.9])
baseline = np.array([0.6, 0.7, 0.65, 0.55, 0.8])

# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(baseline, your_results)
plt.xlabel('AUROC Values using Baseline')
plt.ylabel('AUROC Values using Your Results')
plt.title('Scatter Plot of AUROC for Baseline vs Your Results')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()

# Create a DataFrame for plotting
plot_data = pd.DataFrame({
    'AUROC Values': np.concatenate([baseline, your_results]),
    'Method': ['Baseline'] * len(baseline) + ['Your Results'] * len(your_results)
})

# Violin Plot
plt.figure(figsize=(10, 5))
sns.violinplot(x='Method', y='AUROC Values', data=plot_data)
plt.title('Violin Plot of AUROC Values for Baseline vs Your Results')
plt.show()

# Perform t-test between the two methods
t_stat, p_value = ttest_ind(baseline, your_results)
print(f"T-statistic: {t_stat:.2f}, P-value: {p_value:.2e}")

# Calculate and print the average AUROC values for the two methods
avg_baseline = np.mean(baseline)
avg_your_results = np.mean(your_results)
print(f"Average AUROC Value for Baseline: {avg_baseline:.2f}")
print(f"Average AUROC Value for Your Results: {avg_your_results:.2f}")
```
